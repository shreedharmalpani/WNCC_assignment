```python
a={}
b={}
new_cases={}
cured={}
new_cured={}
deaths={}
new_deaths={}
active={}

count=0
flag= False
search_state= input("Enter the state : ")

for states in states_list:
  state= states.find('span',class_='show-district').text
  total_cases= getattr(states.find('div',class_='td-sc'), 'text', None)

  if(search_state.lower()==state.lower()):

    print(f'{state}      {total_cases}')
    print('\n')
    flag=True
    
    districts_list= states.find_all('div',class_='skgm-tr')
    counter=0
    for districts in districts_list:
      district=districts.find('div',class_='skgm-td').text

      if counter!=0:
        total_cases_districts = getattr(districts.find('div',class_='td-dc'), 'text', None)
        print(f'{district}      {total_cases_districts}')
        
        a[counter]=district
        b[counter]=total_cases_districts

        new_cases[counter]= getattr(districts.find('div',class_='td-ddc'), 'text', None)
        cured[counter]= getattr(districts.find('div',class_='td-dr'), 'text', None)
        new_cured[counter]= getattr(districts.find('div',class_='td-ddr'), 'text', None)
        active[counter]= getattr(districts.find('div',class_='td-da'), 'text', None)
        deaths[counter]= getattr(districts.find('div',class_='td-dd'), 'text', None)
        new_deaths[counter]= getattr(districts.find('div',class_='td-ddd'), 'text', None)

        if new_cases[counter]== '':
          new_cases[counter]='0'
        
        if new_cured[counter]== '':
          new_cured[counter]='0'

        if new_deaths[counter]== '':
          new_deaths[counter]='0'    
        
        



      counter+=1  
      count=counter


      


  if flag== True:
    break

if flag==False:
  print("Either this state doesn't exist or your spelling is wrong ")



 
```

```python
data = {f'Districts in {search_state}': a, ' Total Cases Reported': b}
data
```

```python
import pandas as pd
df = pd.DataFrame(data, index = [i for i in range(1,count)])
df
```

```python
from google.colab import files
df.to_csv(f'wncc_{search_state}.csv') 
files.download(f'wncc_{search_state}.csv')
```
